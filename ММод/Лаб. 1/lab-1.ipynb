```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def mid_square_method(base_num: int, count: int, result_container: [int]):
    tmp = base_num # just for debug output
    orig_len = len(str(base_num))
    base_num = pow(base_num, 2)
    base_num_str = str(base_num)
    while len(base_num_str) < (orig_len * 2):
        base_num_str = "0" + base_num_str
        
    new_len = len(base_num_str)
    offset = (new_len - orig_len) / 2
    while base_num_str[offset] == "0" and offset + orig_len < new_len:
        offset += 1
    print(f"base num: {tmp} base ^2: {base_num_str} offset: {offset}")
    new_num = int(base_num_str[offset:orig_len + offset]) # to always get numbers of the same length 
    result_container.append(new_num)
    count -= 1
    if count == 0:
        return
    else:
        mid_square_method(new_num, count, result_container)
    
```

```python
def conv_to_z(num: int):
    return float("0." + str(num))
    
```

```python
res = []
res_z = []
mid_square_method(19945, 144, res)
for num in res:
    z = conv_to_z(num)
    res_z.append(z)

print(res)
print(res_z)
```

```python
def mul_cong_method(base_a: int, k: int, m: int, a_res: [int], result_container: [int]):
    for i in range(m - 1):
        new_a = (k * base_a) % m # Ai = (k * A(i-1)) mod m
        print(f"{new_a} = ({k} * {base_a}) mod {m}")
        result_container.append(new_a / m)
        a_res.append(new_a)
        base_a = new_a

```

```python

def test_uniformity(K: int, nums: [float]):
    n = nums.count
    interval_size = 1 / K

    intervals = np.zeros(K)
    l = 0
    r = interval_size
    for i in range(K):
        for j in nums:
            if j > l and j < r:
                intervals[i] += 1
        l += interval_size
        r += interval_size
    
    print(intervals)
    
    
    freq = intervals / K

    x = [i for i in np.arange(0.0, 1.0, interval_size)]
   # print(x)

   # print(freq)
   # print(nums)

    plt.hist(x, weights=freq, edgecolor='black')
    
```
